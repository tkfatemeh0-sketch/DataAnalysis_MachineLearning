```python
#outliers 
# quantiles
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1

# boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 = {Q1}")
print(f"Q3 = {Q3}")
print(f"IQR = {IQR}")
print(f"lower boundary = {lower_bound}")
print(f"upper boundary = {upper_bound}")

# remove outliers
df = df[(df['Quantity'] >= lower_bound) & (df['Quantity'] <= upper_bound)]
```

# <font color="696FC7">Predictive Analysis</font>

### <font color="58A0C8">- 1)Predict next month's sales per product</font>

```python
# month column
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.to_period('M')

# new dataset for seperating features and labels
monthly_sales = df.groupby(['Description', 'StockCode', 'Month'])['Quantity'].sum().reset_index()

# months
monthly_sales['Month_num'] = monthly_sales['Month'].apply(lambda x: x.ordinal)

# prediction for a certain product
def predict_next_month(stock_code):
    # that product data
    product_data = monthly_sales[monthly_sales['StockCode'] == stock_code]
    if product_data.empty:
        return f" product with this code {stock_code} not found."

    # regressionmodel
    model = LinearRegression()
    X = product_data[['Month_num']]
    y = product_data['Quantity']
    model.fit(X, y)

    # next month's prediction
    next_month = product_data['Month_num'].max() + 1
    prediction = model.predict([[next_month]])[0]

    return f" Next month's sales prediction (number){stock_code}: {prediction:.2f} "

print(predict_next_month('71459'))  #stockcode
```

### <font color="58A0C8">- 2)Predict return probability for a specific customer</font>

```python
customer_id = '15311.0'
df_customer = df[df['CustomerID'] == customer_id]

# return invoicenumbers 
df_customer['IsReturn'] = df_customer['InvoiceNo'].astype(str).str.startswith('C')

# probabity of return
total_invoices = df_customer['InvoiceNo'].nunique()
return_invoices = df_customer[df_customer['IsReturn']]['InvoiceNo'].nunique()

# zerodevided error
if total_invoices == 0:
    print(f" {customer_id} not exist.")
else:
    prob_return = return_invoices / total_invoices
    print(f"probability of return for customer{customer_id}: {prob_return:.2f}")
```

### <font color="58A0C8">- 3)Highest probability of sales during Christmas</font>

```python
# highest probability of sales in christmas
total_sales = df.groupby('Description')['Quantity'].sum()
df_christmas = df[(df['InvoiceDate'].dt.month == 12) & (df['InvoiceDate'].dt.day >= 15)]
christmas_sales = df_christmas.groupby('Description')['Quantity'].sum()
prob_christmas_sale = (christmas_sales / total_sales).fillna(0).sort_values(ascending=False)
top_christmas_products = prob_christmas_sale.head(10)
top_christmas_products
```

```python
# bar chart
plt.figure(figsize=(8,6))
top_christmas_products.plot(
    kind='bar',
    color='#c62828',
    edgecolor='#2e7d32',
)

plt.title('Top 10 Christmas Hot-Selling Products', fontsize=14, fontweight='bold', color='#2e7d32')
plt.xlabel('Product Description', fontsize=12)
plt.ylabel('Probability of Christmas Sale', fontsize=12)
plt.xticks(rotation=90, ha='right')

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

# <font color="F19ED2">Prescriptive Analysis</font>

### <font color="3AA6B9">- 1)profitable products</font>

```python
# remove returns
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]

# total sales of products
df["TotalSales"] = df["Quantity"] * df["UnitPrice"]
product_sales = df.groupby("Description")["TotalSales"].sum().sort_values(ascending=False)

# top sales products
top_products = product_sales.head(10)

print("Most Profitable Products (Based on Total Sales):")
top_products
```

```python
# product groups
product_sales = (
    df.groupby("Description")["TotalSales"]
    .sum()
    .sort_values(ascending=False)
)
top_products = product_sales.head(10)

# donut chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    top_products,
    labels=top_products.index,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.85
)
centre_circle = plt.Circle((0, 0), 0.70, fc="white")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Most profitable products(based on total sales)", fontsize=14)
plt.tight_layout()
plt.show()
```
